```python
import pandas as pd
from sklearn.decomposition import PCA
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Function to calculate excess returns
def calc_returns(yields, h=12):
    T = len(yields)
    mats = [i * 12 for i in range(1, 11)]  # Maturities from 1 to 10 years (in months)
    mats_h = [m for m in mats if m > h]

    for n in mats_h:
        col_name = f'y{n//12}'
        if col_name in yields.columns:
            if (n - h) in mats:
                col_name_h = f'y{(n-h)//12}'
                yields[f'xr{n//12}'] = (-(n-h) * yields[col_name_h].shift(-h) + 
                                        n * yields[col_name] -
                                        h * yields[f'y{h//12}']) / 12
            else:
                # Approximate: y(t+h)^(n-h) = y(t+h)^(n)
                yields[f'xr{n//12}'] = (-(n-h) * yields[col_name].shift(-h) + 
                                        n * yields[col_name] -
                                        h * yields[f'y{h//12}']) / 12

    xr_columns = [col for col in yields.columns if 'xr' in col]
    yields['xr.avg'] = yields[xr_columns].mean(axis=1, skipna=True)

    return yields.fillna(0)  # Replacing NaN with 0 for simplicity
```

```python
# Load the GSW yields data (example file path, adjust as needed)
gsw_yields_path = 'gsw_yields.csv'
gsw_yields = pd.read_csv(gsw_yields_path)

# Selecting the last observation of each month
gsw_yields['date'] = pd.to_datetime(gsw_yields['date'])
gsw_yields_last_obs = gsw_yields.groupby([gsw_yields['date'].dt.year, gsw_yields['date'].dt.month]).last().reset_index(drop=True)

jps_macro_vars_path = 'jps_macro_vars.csv'  
macro_data = pd.read_csv(jps_macro_vars_path)
macro_data['date'] = pd.to_datetime(macro_data['date'])
```

```python
# Selecting and formatting yields
yield_columns = ['SVENY' + ('0' if i < 10 else '') + str(i) for i in range(1, 11)]
yields = gsw_yields_last_obs[yield_columns] / 100  # Divide by 100 as per script
yields.columns = ['y' + str(i) for i in range(1, 11)]  # Renaming columns to y1 to y10

# Constructing regression data
reg_data = calc_returns(yields)

# Adding dates and calculating the slope
reg_data['date'] = gsw_yields_last_obs['date'].reset_index(drop=True)
reg_data['Slope'] = reg_data['y10'] - reg_data['y1']

# Calculating principal components
pca_columns = ['y' + str(i) for i in range(1, 11)]
pca = PCA(n_components=3)
principal_components = pca.fit_transform(reg_data[pca_columns])

reg_data['PC1'] = principal_components[:, 0]
reg_data['PC2'] = -principal_components[:, 1]
reg_data['PC3'] = -principal_components[:, 2]
reg_data = reg_data[reg_data['date'] > pd.Timestamp('1984-12-31')]
```

```python
# Extract year and month from the date columns
reg_data['year'] = reg_data['date'].dt.year
reg_data['month'] = reg_data['date'].dt.month
macro_data['year'] = macro_data['date'].dt.year
macro_data['month'] = macro_data['date'].dt.month

# Merge the datasets on year and month
# Dropping the 'date' column from macro_data before merging
macro_data_no_date = macro_data.drop(columns='date')

# Merging reg_data with macro_data_no_date
reg_data = pd.merge(reg_data, macro_data_no_date, on=['year', 'month'], how='left')

# Optionally, you can drop the extra year and month columns if they're no longer needed
reg_data.drop(columns=['year', 'month'], inplace=True)
```

```python
# Plotting Slope and Average Excess Returns (xr.avg) over time
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plotting Slope with blue color on the left y-axis
ax1.plot(reg_data['date'], reg_data['Slope'], label='Slope', color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('Slope', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Creating a second y-axis for Average Excess Returns
ax2 = ax1.twinx()
ax2.plot(reg_data['date'], reg_data['xr.avg'], label='Average Excess Returns (xr.avg)', color='green')
ax2.set_ylabel('Average Excess Returns (xr.avg)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

plt.title('Slope and Average Excess Returns Over Time')
plt.grid(True)
plt.show()



# plot principal components --------------------------
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plotting Slope with blue color on the left y-axis
ax1.plot(reg_data['date'], reg_data['PC1'], label='PC1', color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('PC1', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Creating a second y-axis for Average Excess Returns
ax2 = ax1.twinx()
ax2.plot(reg_data['date'], reg_data['PC2'], label='PC2', color='green')
ax2.set_ylabel('PC2', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Creating a second y-axis for Average Excess Returns
ax2 = ax1.twinx()
ax2.plot(reg_data['date'], reg_data['PC3'], label='PC3', color='red')
ax2.set_ylabel('PC3', color='green')
ax2.tick_params(axis='y', labelcolor='green')

plt.title('Principal Components')
plt.grid(True)
plt.show()

# plot for macro variables --------------------------
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plotting Slope with blue color on the left y-axis
ax1.plot(reg_data['date'], reg_data['GRO'], label='Growth', color='blue')
ax1.set_xlabel('Date')
ax1.set_ylabel('Growth', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Creating a second y-axis for Average Excess Returns
ax2 = ax1.twinx()
ax2.plot(reg_data['date'], reg_data['INF'], label='Infaltion', color='green')
ax2.set_ylabel('Inflation', color='green')
ax2.tick_params(axis='y', labelcolor='green')

plt.title('Macro Variables')
plt.grid(True)
plt.show()
```

# Run CS regressions to test EH

```python
# Independent variable (Slope) and dependent variable (xr.avg)
X = reg_data['Slope']
y = reg_data['xr.avg']

# Adding a constant to the independent variable
X = sm.add_constant(X)

# Performing the regression
cs_ols = sm.OLS(y, X).fit()

# Computing standard errors using Newey-West estimator
se_cs = np.sqrt(np.diag(cs_ols.get_robustcov_results(cov_type='HAC', maxlags=18).cov_params()))

# Computing t-statistics
t_cs = cs_ols.params / se_cs

# Creating a summary dataframe for the regression results
reg_results = cs_ols.summary2().tables[1]
reg_results['std.error'] = se_cs
reg_results['t'] = t_cs

# Adding confidence intervals
conf_int = cs_ols.conf_int()
reg_results['[0.025'] = conf_int[0]
reg_results['0.975]'] = conf_int[1]

# Displaying the regression summary
print(reg_results)
```

```python
# Independent variable (Slope) and dependent variable (xr.avg)
X_pc = reg_data[['PC1', 'PC2', 'PC3']]
y = reg_data['xr.avg']

# Adding a constant to the independent variable
X_pc = sm.add_constant(X_pc)

# Performing the regression
pc_ols = sm.OLS(y, X_pc).fit()

# Computing standard errors using Newey-West estimator
se_pc = np.sqrt(np.diag(pc_ols.get_robustcov_results(cov_type='HAC', maxlags=18).cov_params()))

# Computing t-statistics
t_pc = pc_ols.params / se_pc

# Creating a summary dataframe for the regression results
reg_results = pc_ols.summary2().tables[1]
reg_results['std.error'] = se_pc
reg_results['t'] = t_pc

# Adding confidence intervals
conf_int = cs_ols.conf_int()
reg_results['[0.025'] = conf_int[0]
reg_results['0.975]'] = conf_int[1]

# Displaying the regression summary
print(reg_results)
```

```python
# compute mpsp := 1/n sum_i (yi - hat{y})^2
pc_ols_predictions = pc_ols.predict(X_pc)
mspe_pc_ols = ((pc_ols_predictions - y) ** 2).mean()
```

```python

# Preparing the data
X = reg_data[['PC1', 'PC2', 'PC3']]
y = reg_data['xr.avg']

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Creating and training the Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Making predictions on the test set
predictions_rf = rf_model.predict(X_test)

# Calculating MSPE for the Random Forest model
mspe_rf = mean_squared_error(y_test, predictions_rf)
```

```python
# Comparison of MSPE
print(f"MSPE of Linear Model (PCs): {mspe_pc_ols:.4f}")
print(f"MSPE of Random Forest Model (PCs): {mspe_rf:.4f}")
print(f"Improvement from RF: {100*(mspe_pc_ols-mspe_rf)/mspe_pc_ols:.4f}%")
```
