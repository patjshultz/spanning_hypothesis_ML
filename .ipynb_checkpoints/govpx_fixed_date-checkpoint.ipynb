```python
import wrds
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def price_to_yield_semi_annual(price, coupon, tau):
    """
    Converts bond price to yield for semi-annual coupon bonds.
    :param price: Bond price
    :param coupon: Annual coupon rate of the bond
    :param tau: Time to maturity in years
    :return: Yield for semi-annual coupon bond
    """
    if tau <= 0 or price <= 0:
        return float('nan')  # Return NaN for non-feasible values
    
    periods = int(tau * 2)  # Number of six-month periods
    coupon_payment = coupon / 2  # Semi-annual coupon payment
    
    # Initial guess for YTM
    ytm = coupon / price
    tolerance = 1e-6
    max_iterations = 1000
    iteration = 0

    while iteration < max_iterations:
        price_calculated = sum([coupon_payment / (1 + ytm / 2) ** (i + 1) for i in range(periods - 1)]) + (100 + coupon_payment) / (1 + ytm / 2) ** periods
        diff = price_calculated - price
        
        if abs(diff) < tolerance:
            break

        ytm -= diff / 10  # Adjust the YTM guess
        if ytm < 0:
            ytm = tolerance  # Prevent negative YTM

        iteration += 1

    if iteration == max_iterations:
        return float('nan')  # Return NaN if no convergence

    return ytm * 100  # Convert to percentage

# Example usage: 
# Assuming you have the row data as before
# yield_result = price_to_yield_semi_annual(row['midprice'], row['coupon'], row['tau'])
# print(yield_result)

def subset_and_filter_data(dataframe, start_minute):
    """
    For each cusip, selects the row that has the time closest to the start time, 
    sorts the result by tau, and drops observations with tau > 10.

    :param dataframe: pandas DataFrame containing the data.
    :param start_minute: Integer representing the starting minute.
    :return: Processed DataFrame.
    """
    end_minute = start_minute + 1

    # Filter the dataframe for the specified time range
    filtered_df = dataframe[(dataframe['minute'] >= start_minute) & (dataframe['minute'] < end_minute)]

    # Group by 'cusip' and select the row with the closest time to start_minute
    closest_rows = filtered_df.groupby('cusip').apply(lambda x: x.loc[(x['minute'] - start_minute).abs().idxmin()])

    # Sort the resulting DataFrame by 'tau'
    sorted_closest_rows = closest_rows.sort_values(by='tau')

    # Drop observations where tau > 10
    final_df = sorted_closest_rows[sorted_closest_rows['tau'] <= 10].reset_index(drop=True)
    return_df = final_df[final_df['tau'] >= 0].reset_index(drop=True)
    
    return return_df

import numpy as np

def bootstrap_yield_curve(df):
    zero_coupon_yields = {}
    
    for index, row in df.iterrows():
        tau, ytm, coupon = row['tau'], row['ytm'], row['coupon']
        
        # Handle short maturity bonds (less than 6 months) as zero-coupon bonds
        if tau < 0.5:
            zero_coupon_yields[tau] = ytm  # Directly use YTM for bonds with maturities less than 6 months
            continue

        # Calculate cash flows for semi-annual coupon bonds
        num_payments = int(np.ceil(tau * 2))
        cash_flows = np.full(num_payments, coupon / 2)
        cash_flows[-1] += 100  # Add the face value to the last payment

        # Calculate the present value of cash flows using the yields for shorter maturities
        present_value = 0
        for i in range(num_payments - 1):
            if (i + 1) / 2 in zero_coupon_yields:
                discount_factor = np.exp(-zero_coupon_yields[(i + 1) / 2] * ((i + 1) / 2))
                present_value += cash_flows[i] * discount_factor
            else:
                # If we don't have the yield for a shorter maturity, use linear interpolation or extrapolation
                # Here, as a simplification, we'll just break and use the YTM as an approximation
                present_value = 0
                break

        # Calculate the yield for the remaining cash flow
        if present_value > 0:
            remaining_cash_flow = cash_flows[-1]
            remaining_time = tau - (num_payments - 1) / 2
            # Calculate yield using the remaining cash flow and present value
            zero_coupon_yield = -np.log(remaining_cash_flow / (100 - present_value)) / remaining_time
            zero_coupon_yields[tau] = zero_coupon_yield
        else:
            # If no present value was calculated, approximate with YTM
            zero_coupon_yields[tau] = ytm

    return zero_coupon_yields

# Example usage
# zero_coupon_curve = bootstrap_yield_curve(df)
# print(zero_coupon_curve)

def plot_yield_curve(yield_curve):
    """
    Plots the zero-coupon yield curve.

    :param yield_curve: Dictionary with maturities as keys and yields as values.
    """
    # Sort the yield curve by maturity
    sorted_yield_curve = dict(sorted(yield_curve.items()))

    # Extract maturities and yields
    maturities = list(sorted_yield_curve.keys())
    yields = list(sorted_yield_curve.values())

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(maturities, yields, marker='o')
    plt.title("Zero-Coupon Yield Curve")
    plt.xlabel("Time to Maturity (Years)")
    plt.ylabel("Yield (%)")
    plt.grid(True)
    plt.show()


def nelson_siegel_svensson(tau, beta0, beta1, beta2, beta3, lambda1, lambda2):
    return (beta0 + 
            beta1 * (1 - np.exp(-tau / lambda1)) / (tau / lambda1) + 
            beta2 * ((1 - np.exp(-tau / lambda1)) / (tau / lambda1) - np.exp(-tau / lambda1)) +
            beta3 * ((1 - np.exp(-tau / lambda2)) / (tau / lambda2) - np.exp(-tau / lambda2)))
```

```python
# List all databases
all_databases = db.list_libraries()

# Filter databases starting with 'govpx'
govpx_databases = [db for db in all_databases if db.startswith('govpx')]

print(govpx_databases)
```

```python

# Replace 'your_username' with your WRDS username
# You will be prompted to enter your password

# Querying the GovPX dataset
# Modify the SQL query according to your specific requirements

govpx_data = db.raw_sql("""
SELECT cusip, coupon, matdate, type, indbid, indask, indbyld, indayld, date, time
FROM govpx_ticks2022.ticks22
WHERE date = '2022-02-23'
""")

# For example, to see the first few rows of the dataset:
print(govpx_data.head())
print("Number of rows in the DataFrame:", govpx_data.shape[0])
```

```python
# clean daily data
# List of columns to check for NaN values
columns_to_check = ['indbid', 'indbyld', 'indask', 'indayld']

# Drop rows where all specified columns are NaN
govpx_data_cleaned = govpx_data.dropna(subset=columns_to_check, how='all')
print("Number of rows in the cleaned DataFrame:", govpx_data_cleaned.shape[0])
```

```python
govpx_data_cleaned.tail()
```

```python
# drop on the run securities
govpx_data_filtered = govpx_data_cleaned[govpx_data_cleaned['type'] != 152]
print("Number of rows in the cleaned DataFrame:", govpx_data_filtered.shape[0])
```

```python
# Create a copy of the DataFrame slice to avoid SettingWithCopyWarning
govpx_data_filtered = govpx_data_filtered.copy()

# Now, you can safely make modifications
govpx_data_filtered.loc[:, 'midprice'] = (govpx_data_filtered.loc[:, 'indbid'] + govpx_data_filtered.loc[:, 'indask']) / 2
govpx_data_filtered.loc[:, 'tau'] = (pd.to_datetime(govpx_data_filtered.loc[:, 'matdate']) - pd.to_datetime(govpx_data_filtered.loc[:, 'date'])).dt.days / 365
govpx_data_filtered.loc[:, 'ytm'] = (govpx_data_filtered.loc[:, 'indbyld'] + govpx_data_filtered.loc[:, 'indayld']) / 2
govpx_data_filtered.loc[:, 'minute'] = govpx_data_filtered.loc[:, 'time'] / 60
```

```python
processed_data = subset_and_filter_data(govpx_data_filtered, 1250)

# bootstrap a zcy
zero_coupon_curve = bootstrap_yield_curve(processed_data)

# Assuming 'maturities' and 'yields' are your data from the bootstrapped yield curve
maturities = np.array(list(zero_coupon_curve.keys()))
yields = np.array(list(zero_coupon_curve.values()))

initial_guess = [0.03, -0.02, 0.02, -0.01, 1.0, 2.0]  # Initial guess for the parameters
params, params_covariance = curve_fit(nelson_siegel_svensson, maturities, yields, p0=initial_guess)

# Plotting the fitted curve
plt.figure(figsize=(10, 6))
plt.scatter(maturities, yields, label='Observed Data')
fitted_yields = nelson_siegel_svensson(maturities, *params)
plt.plot(maturities, fitted_yields, label='Fitted Nelson-Siegel-Svensson Curve', color='red')
plt.title("Yield Curve Fitted with Nelson-Siegel-Svensson Model")
plt.xlabel("Maturity (Years)")
plt.ylabel("Yield")
plt.legend()
plt.show()
```
